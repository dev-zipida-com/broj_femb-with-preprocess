```python
# JSON   
def find_json_files(folder):
    return [file for file in os.listdir(folder) if file.endswith('.json')]

#   JSON  
all_json_files = []
with ThreadPoolExecutor() as executor:
    results = executor.map(find_json_files, folder_list)
    for json_files in results:
        all_json_files.extend(json_files)

#  JSON  
output_file = 'collected_json_files.json'
with open(output_file, 'w') as f:
    json.dump(all_json_files, f)

print(f" JSON   '{output_file}' .")
```

```python
all_json_files = []
for folder in folder_list:
    json_files = [file for file in os.listdir(folder) if file.endswith('.json')]
    all_json_files.extend(json_files)
ouput_file = 'collected_json_files.json'
with open(ouput_file, 'w') as f:
    json.dump(all_json_files, f)
```
